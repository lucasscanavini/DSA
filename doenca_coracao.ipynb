```python
# carregando os pacotes
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.datasets import fetch_openml              # dataset
from sklearn.model_selection import cross_val_score
```

```python
# carregando os dados
dados = fetch_openml('heart')
```

```python
dados
```

```python
x = dados.data
y = np.copy(dados.target)
y[y == -1] = 0
```

```python
# dividindo os dados de treio e teste
x_treino, y_treino = x[189:], y[189:]
x_teste, y_teste = x[:189], y[:189]
```

```python
# construindo o estimador
# https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html

estimador = DecisionTreeClassifier(max_depth= 1, min_samples_leaf= 1)
```

```python
# http://scikit-learn.org/stable/modules/generated/sklearn.ensemble.AdaBoostClassifier.html

Ada_boost = AdaBoostClassifier(base_estimator= estimador,
                                learning_rate= 0.1,
                                n_estimators= 400,
                                algorithm= 'SAMME')
```

```python
# treinando o modelo
Ada_boost.fit(x_treino, y_treino)
```

```python
# verificando o score
scores = cross_val_score(Ada_boost, x_teste, y_teste)
media_score = scores.mean()
print(media_score) 
```
